```python
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1
outliers=((df<(Q1-1.5*IQR))  | (df>(Q3+1.5*IQR))).sum()
print(outliers)
```

## 2.Statistical Summary

```python
numeric_df=df.select_dtypes(include='number')
numeric_df
```

```python
summary = pd.DataFrame({
    'Mean': numeric_df.mean(),
    'Median': numeric_df.median(),
    'Std Dev': numeric_df.std(),
    'Q1 (25%)': numeric_df.quantile(0.25),
    'Q3 (75%)': numeric_df.quantile(0.75),
    'IQR': numeric_df.quantile(0.75) - numeric_df.quantile(0.25)
})
```

```python
summary
```

## 3.Data Visualization

#### Histogram 

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
numeric_df.hist(figsize=(15, 12),bins=30,color='skyblue',edgecolor='black')
plt.suptitle("Histograms of Numerical Variables", fontsize=16)
plt.tight_layout()
plt.show()
```

#### Boxplot

```python
plt.figure(figsize=(15, 8))
sns.boxplot(data=numeric_df)
plt.xticks(rotation=45)
plt.title("Boxplots of Numerical Features")
plt.show()
```

#### Barchart

```python
if 'NSP' in df.columns:
    plt.figure(figsize=(6, 6))
    df['NSP'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))
    plt.title("Distribution of Target Variable (NSP)")
    plt.ylabel("")
    plt.show()
```

#### Pie Chart

```python
 sns.countplot(x='NSP', data=df, palette='pastel')
 plt.title("Count of Each Class in NSP")
 plt.show()
```

#### Scatter plot

```python
plt.figure(figsize=(12, 10))
corr = numeric_df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

#### Correlation Heatmap

```python
if 'LB' in df.columns and 'ASTV' in df.columns:
    sns.scatterplot(x='LB', y='ASTV', data=df)
    plt.title("Scatter Plot: LB vs ASTV")
    plt.show()
```

#### pairplot

```python
# Pair Plot (lightly sampled for speed if dataset is large)
sns.pairplot(numeric_df.sample(200), diag_kind='kde')
plt.suptitle("Pair Plot of Numerical Features", y=1.02)
plt.show()
```

```python
if 'NSP' in df.columns and 'ASTV' in df.columns:
    plt.figure(figsize=(8, 6))
    sns.violinplot(x='NSP', y='ASTV', data=df, palette='pastel')
    plt.title("Violin Plot: ASTV by NSP")
    plt.show()
```

## 4.Pattern Recognization and insights

```python
correlation_matrix = df.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```
